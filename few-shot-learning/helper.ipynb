```python
import numpy as np
from scipy import misc
import os
import cv2
from PIL import Image

dataset = []
examples = []

base_dir = './Covid19-dataset/'

data_root = os.path.join(base_dir, 'test')
alphabets = os.listdir(data_root)
for alphabet in alphabets:
    if alphabet in '.DS_Store':
        continue
    characters = os.listdir(os.path.join(data_root, alphabet))
    for img_file in characters:
        if img_file in '.DS_Store':
            print (".DS_Store")
        else:
            img = cv2.imread(os.path.join(data_root, alphabet, img_file), cv2.IMREAD_COLOR)
            img = cv2.resize(img, (32,32))
#             img = np.stack((img,)*3, axis=-1)
            print (img.shape)
            examples.append(img)
    dataset.append(examples)
print (np.asarray(dataset).shape)
t = np.asarray(dataset)
# t = t.reshape(t.shape[1], t.shape[0], t.shape[2], t.shape[3], t.shape[4])
print (t.shape)
np.save("./data_32_test_covid_3.npy", np.asarray(t))
```
